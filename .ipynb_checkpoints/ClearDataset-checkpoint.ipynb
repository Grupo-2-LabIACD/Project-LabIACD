# Clear Dataset

Neste notebook vamos limpar o data set formado pelo pyradiomcs para nos tentar livrar de features inuteis e colunas que não tem informação util

```python
import pandas as pd
import numpy as np
import os
```

## Load dataset 

```python
original_dataset=pd.read_csv('dataset.csv')
original_dataset_copy=original_dataset
pd.set_option('display.max_columns', None) 
original_dataset
```

## Findings

Looking at the dataset we can see a series of problems:<br>
- Patient_id, is an identifier and not a feature so feading this to our models will only create problems.<br>
- Nodule_Id, is also an identifier and not a feature and although it might be possible to feed it to the models it will cause severe problems, and after if we need to identify the nodules again due to the dataset beeing ordered by nodule id we'll know that line = 0 will be nodule = 1 and so on.<br>
- Scan_id, once again an identifier<br>
- Internalstructure, at first sight it seems like the value is always 1.0 so it's useless<br>
- Malignancy is of float type, and beeing that we are dealing with a classification and not regression problem we'll have to round to the nearest int.<br>

## Solving Problems

### Patient_id

```python
original_dataset_copy=original_dataset_copy.drop('Patient_id', axis=1)
original_dataset_copy
```

### Nodule_id

```python
original_dataset_copy=original_dataset_copy.drop('Nodule_Id', axis=1)
original_dataset_copy
```

### Scan_Id

```python
original_dataset_copy=original_dataset_copy.drop('Scan_id', axis=1)
original_dataset_copy
```

### Internal structure

#### See if there are diferences

```python
column_internal=original_dataset['Internalstructure']
i=1
while (column_internal[i] == column_internal[0]):
    i+=1
else:
    print(f'Value change detected in index = {i} value = {column_internal[i]}.')
```

As we can see what we thought at first sight to be a useless column it actualy has value changes so we'll keep it

### Malignancy

```python
original_dataset_copy['Malignancy'] = original_dataset['Malignancy'].round(0)
original_dataset_copy
```

## Export dataset

```python
original_dataset_copy.to_csv('Dataset_cleared.csv')
check=os.path.isfile('Dataset_cleared.csv')
print(f'Csv saved = {check}.')
```
